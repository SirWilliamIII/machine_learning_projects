# Boston Housing Analysis

- Student-teacher ratio (PTRATIO)
- Proportion of African American by town (AA)
- Median value of owner-occupied homes x1000 dollars (MEDV)

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('housing.data', header=None, delim_whitespace=True)
df.head()
```

```python
cols = ['CRIM', 'ZN', 'INDUS', 'CHAS', 'NOX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'AA', 'LSTAT', 'MEDV']
df.columns = cols
df.head()
```

```python
cols_study = ['PTRATIO', 'AA', 'MEDV']
sns.pairplot(df[cols_study], size=3)
plt.show()
```
